```python
import matplotlib.pyplot as plt
import numpy as np


def plot_bomba(vetor):
    n = len(vetor) // 2
    horas = vetor[:n]
    duracoes = vetor[n:]
    
    # Criando os intervalos de funcionamento
    inicio = np.array(horas)
    fim = inicio + np.array(duracoes)
    
    # Criando o grfico
    plt.figure(figsize=(10, 5))
    for i in range(n):
        plt.plot([inicio[i], fim[i]], [1, 1], linewidth=5, label=f'Ciclo {i+1}' if i == 0 else "")
    
    plt.xlabel('Hora')
    plt.ylabel('Funcionamento')
    plt.title('Grfico de Funcionamento da Bomba')
    plt.yticks([])  # Removendo valores do eixo Y
    plt.xticks(np.arange(0, 24.5, 0.5))  # Definindo marcaes de meia em meia hora
    plt.legend()
    plt.grid(True, axis='x', linestyle='--', alpha=0.7)
    plt.show()
```
